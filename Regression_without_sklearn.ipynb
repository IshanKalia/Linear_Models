## Loading dataset

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 

penguins = pd.read_csv("/Users/ishankalia/Documents/GitHub/Linear_Models/penguins_lter.csv")
feature_name = "Flipper Length (mm)"
target_name = "Body Mass (g)"
data, target = penguins[feature_name], penguins[target_name]
penguins.head(5)
```

## Model Construction 

```python
#Constructing a linear model function
def linear_model(flipper_length, weight_flipper_length, intercept_body_mass):
    body_mass = weight_flipper_length * flipper_length + intercept_body_mass
    return body_mass
```

```python
# Construct a vector weights and vector intercepts of the same length 
flipper_length_range = np.linspace(data.min(), data.max(), num = 300)
weights = np.ndarray(shape = (5,1))
intercepts = np.ndarray(shape = (5,1))
weights[0] = 45
weights[1] = 50
weights[2] = 55
weights[3] = 40
weights[4] = 35

intercepts[0] = -5000
intercepts[1] = -5500
intercepts[2] = -6000
intercepts[3] = -4500
intercepts[4] = -4000

ax = sns.scatterplot(data = penguins, x = data, y = target)
ax.set_title("Scatter Plot of the Flipper Length and the Body Mass")

#Zipping the parameters together:
parameters = list(zip(weights, intercepts))
for weight, intercept in parameters: 
    body_mass = linear_model(flipper_length_range, weight, intercept)
    ax.plot(flipper_length_range, body_mass)

plt.show()
```
